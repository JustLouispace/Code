## 4. Pie charts

#### A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. In a pie chart, the arc length of each slice is proportional to the quantity it represents.

```python
import matplotlib.pyplot as plt
import seaborn

# declare data
data = [60, 22, 40, 41, 32]
keys = ['Finance', 'Accounting', 'Psychology', 'Marketing', 'Laws']

# define Seaborn color palette to use
palette_color = seaborn.color_palette('Set2')  # Set2, paster

# plotting data on chart
plt.pie(data, labels=keys, colors=palette_color,
        explode=[0, 0, 0, 0, 0.3], autopct='%.0f%%')

# displaying chart
plt.title('Students taking elective courses')
plt.show()
```

## 5. Distribution plots
While histogram can visualize distribution of one variable, seaborn offers several different approaches to visualizing the multivariate distributions of data.

### Recall of matplotlib's histogram

```python
plt.hist(df['tip'], bins=20)    # The distribution plot of tip by matplotlib
```

## Seaborn Distribution plots


#### Flexibly plot a univariate distribution of observations. 
This function provides access to several approaches for visualizing the univariate or bivariate distribution of data. The <font color='blue'>kind</font> parameter selects the approach to use:

<li>histplot() (with kind="hist"; the default) A hitogram plot of a univariate data.

<li>kdeplot() (with kde=True) A kernel density estimate (KDE) plot for visualizing the distribution of observations in a dataset, analagous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

<li>ecdfplot() (with kind="ecdf") Empirical cumulative distribition function


### Histogram

```python
# histogram with seaborn's displot()
#
sns.displot(df['tip'], bins=20)    # The histogram (default kind) plot of tip with seaborn
```

### KDE (Kernel Density Estimate) plot
KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate:

```python
# Kernel Density Estimate plot
#
sns.displot(df['tip'], kind='kde')
```

```python
# Kernel Density Estimate plot, with hue=smoker grouping
#
sns.displot(df, x='tip', hue='smoker', kind='kde') # , fill=True)
```

```python
# kde plot with stacking smoker grouping (like area chart)
#
sns.displot(df, x='tip', hue='smoker', kind='kde', multiple='stack')
```

```python
# histogram together with kernel density estimate and mean value
#
sns.displot(df['tip'], bins=20, kde=True) # histogram with KDE

# add axis verticle line (axvline)
#
plt.axvline(df['tip'].mean(), 0.3, 0.7, color='red')  # ploting the mean value in red
```

### ECDF (Empirical Cumulative Distribution Function) plot
ECDF plot an estimate of the cumulative distribution function that generated the points in the sample. It smooths an increasing curve through each datapoint such that the height of the curve reflects the proportion of observations with a smaller value:

```python
# Empirical Cumulative Distribution Function (ฟังก์ชันการกระจายสะสมค่าจากการทดลอง)
#
sns.displot(df['tip'], kind='ecdf')
```

```python
# Empirical Cumulative Distribution Function with smoker grouping
#
sns.displot(df, x='tip', kind='ecdf', hue='sex')
```

## 12. FacetGrid

FacetGrid visualizes the distribution of a variable or the relationship between multiple variables separately within subsets of your dataset.
<li>
    The variables should be categorical and the data at each level of the variable will be used for a facet along that axis.
 <li>A FacetGrid can be drawn with up to three dimensions: row, col, and hue.

```python
# univariate, use time to differentiate by col
#
g = sns.FacetGrid(df, col='time')
g.map(sns.histplot, 'tip')
```

```python
# bivariate, use sex and smoker to differentiate cols and hues
#
g = sns.FacetGrid(df, col='sex', hue='smoker')
g.map(sns.scatterplot, 'total_bill', 'tip', alpha=.7)
plt.figure(figsize=(12,7))
g.add_legend()
```

```python
# enlarge the figure and font size
#
sns.set(font_scale=1.5)
g = sns.FacetGrid(df, col='sex', hue='smoker', height=6)
g.map(sns.scatterplot, 'total_bill', 'tip', alpha=.7)
plt.figure(figsize=(12,7))
g.add_legend()
```
