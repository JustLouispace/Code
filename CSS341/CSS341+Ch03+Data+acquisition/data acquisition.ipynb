```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
codeall_ = []
descall_ = []
areaall_ = []
priceall_ = []
imageall_ = []
linkall_ = []
```

```python
for i in range(1,37): 
    webpage = requests.get('https://www.ttbbank.com/property/property/for-sale?type=&price=&province=10&amphur=&area=&limit=6&page='+str(i))
    soures = BeautifulSoup(webpage.text, 'html')
    properties = soures.find_all('div', class_='col-list-items')
    for prop in properties:
        code = prop.find('div', class_='entry-code').text
        codeall_.append(code)
        desc = prop.find('p', class_='entry-desc').text
        descall_.append(desc)
        area = prop.find('p', class_='entry-area').text
        areaall_.append(area)
        price = prop.find('p', class_='entry-price').text
        priceall_.append(price)
        image = prop.find('img').attrs['src']
        imageall_.append(image)
        link = prop.find('a').attrs['href']
        linkall_.append(link)
```

```python
data = {'Code':[],'Location':[],'Area':[],'Price':[],'Image':[],'Link':[]}
data['Code'] = codeall_
data['Location'] = descall_
data['Area'] = areaall_
data['Price'] = priceall_
data['Image'] = imageall_
data['Link'] = linkall_
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_csv('ttbbank.csv', index=False)
```
